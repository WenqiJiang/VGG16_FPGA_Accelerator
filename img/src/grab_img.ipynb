# Collect 1000 jpg images

## Download images

```python
urls = []

count = 0
with open("fall11_urls.txt") as f:
    for line in f:
        if line.split("\t")[1][-4:-1] == 'jpg':
            urls.append((line.split("\t")[0], line.split("\t")[1][:-1]))
            count += 1
        
        if count >= 1000:
            break
            
print(urls[0])
```

```python
print(len(urls))
print(urls)
```

```python
for i, ele in enumerate(urls):
    url = ele[1]
    img_name = "{}.jpg".format(i)
    !wget $url -O $img_name --timeout=2 --tries=1
```

## move to folder & clean out empty images

```python
mv *.jpg ../origin_img/
```

```python
import os
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
  

max_size = 0
min_size = 10000000
count = 0
for i in range(1000):
    try:
        statinfo = os.stat('../origin_img/{}.jpg'.format(i))
#         img = mpimg.imread('../origin_img/{}.jpg'.format(i))
#         print(img.size)
        if statinfo.st_size > 40000:
            count += 1
            copy_from = '../origin_img/{}.jpg'.format(i)
            copy_to = '../clean_img/{}.jpg'.format(count)
            !cp $copy_from $copy_to 
        if statinfo.st_size > max_size:
            max_size = statinfo.st_size
#         print(statinfo.st_size)
    except:
        pass
    
print("max size: {}\t count: {}".format(max_size, count))
print(min_size)
```

### then manually delete damaged images...

```python
count = 0

for i in range(1000):
    try:
        statinfo = os.stat('../clean_img/{}.jpg'.format(i))
        count += 1
        name = '../clean_img/{}.jpg'.format(i)
        rename = '../clean_img/{}.jpg'.format(count)
        !mv $name $rename
    except:
        pass
    
print("count: {}".format(count))
```

## Resize images & save BMP

```python
from PIL import Image
from resizeimage import resizeimage

for i in range(1, count + 1):

    with open('../clean_img/{}.jpg'.format(i), 'r+b') as f:
        with Image.open(f) as image:
            cover = resizeimage.resize_cover(image, [224, 224])
            # save to BMP -> that is the uncompressed version of image
            print(cover.size)
            cover.save('../resize_img/{}.bmp'.format(i), 'BMP')
    
```
